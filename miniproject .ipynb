```python
def registration():
    employee_id = int(input("Enter Employee Id: "))
    name = input("Enter Name: ")
    age = int(input("Enter age: "))
    department = input("Enter Department: ")
    salary = float(input("Enter salary: "))

    if age >= 18 and salary > 5000:
        print("Employee registered successfully")
        return employee_id, salary
    else:
        print("Age must be >= 18 and salary > 0")
        return None, None


def attendance():
    attend = []

    for i in range(5):
        at = input("Enter attendance : ")
        if at == "p":
            attend.append("present")
        else:
            attend.append("absent")

    present_days = attend.count("present")
    print("Attendance record:", attend)

    if present_days >= 4:
        print("Regular")
    elif present_days == 3:
        print("Warning")
    else:
        print("Irregular")

    return present_days


def salary_calculation(basic_salary, present_days):
    hra = basic_salary * 0.20
    da = basic_salary * 0.10
    fd = basic_salary * 0.12

    net_salary = (basic_salary + hra + da) - fd

    if present_days < 3:
        print("Salary after deduction:", basic_salary - fd)
    else:
        print("Net Salary:", net_salary)


def rating():
    rate = int(input("Enter your rating: "))
    if rate >= 4:
        print("Excellent")
    elif rate >= 3:
        print("Good")
    else:
        print("Need to improve")


def login(correct_user_id, correct_password):
    user = int(input("Enter your ID: "))
    pwd = input("Enter your password: ")

    if user == correct_user_id and pwd == correct_password:
        print("Login successful!!")
        return True
    else:
        print("User ID or password is incorrect!!")
        return False



user_id, salary = registration()

if user_id is not None:
    if login(user_id, "admin123"):
        present_days = attendance()
        salary_calculation(salary, present_days)
        print("Final feedback of employee")
        rating()
```
